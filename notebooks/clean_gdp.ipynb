```python
import pandas as pd
import numpy as np
```

```python
gdp_raw = pd.read_csv("../data/raw/SAGDP/SAGDP2__ALL_AREAS_1997_2024.csv")
```

```python
gdp = gdp_raw.copy()
```

```python
gdp.head()
```

```python
gdp[gdp["Description"] == "All industry total "]
```

```python
gdp["Description"] = gdp["Description"].str.strip()
```

```python
gdp = gdp[gdp["Description"] == "All industry total"]
```

```python
gdp.head()
```

```python
year_cols = [col for col in gdp.columns if col.isdigit()]
gdp = gdp[["GeoName"] + year_cols]
```

```python
gdp.head()
```

```python
gdp = gdp.rename(columns={"GeoName": "State"})
```

```python
gdp.head()
```

```python
gdp = gdp[gdp["State"] != "United States *"]
```

```python
gdp.head()
```

```python
gdp = gdp.reset_index(drop=True)
```

```python
gdp.head()
```

```python
gdp = gdp.melt(id_vars="State", var_name="Year", value_name="GDP")
```

```python
gdp.head()
```

```python
gdp["Year"] = gdp["Year"].astype(int)
gdp["GDP"] = pd.to_numeric(gdp["GDP"], errors="coerce")
```

```python
gdp.info()
```

```python
gdp.head()
```

```python
gdp = gdp.sort_values(by=["State", "Year"])
```

```python
gdp.head(30)
```

```python
gdp = gdp.reset_index(drop=True)
```

```python
gdp.head()
```

```python
gdp.shape
```

```python
gdp.columns
```

```python
gdp.info()
```

```python
gdp["State"].unique()
```

```python
regions = [
    "Far West", "Great Lakes", "Mideast", "New England",
    "Plains", "Rocky Mountain", "Southeast", "Southwest"
]

gdp = gdp[~gdp["State"].isin(regions)]
```

```python
gdp["State"].unique()
```

```python
gdp.head(20)
```

```python
gdp.to_csv("../data/cleaned/gdp_clean.csv", index=False)
```
