```python
import pandas as pd
import numpy as np
```

```python
df_employment_raw = pd.read_csv("../data/raw/employment.csv")
df_employment = df_employment_raw.copy()
```

```python
df_employment.head()
```

```python
df_employment = df_employment[df_employment["GeoName"] != "United States"]
```

```python
df_employment.head()
```

```python
df_employment = df_employment.drop(columns="GeoFIPS")
```

```python
df_employment.info()
```

```python
for i in range(1998, 2008):
    df_employment = df_employment.drop(columns=str(i))
```

```python
df_employment.head()
```

```python
df_employment = df_employment.rename(columns={"GeoName":"State"})
```

```python
df_employment = df_employment.melt(id_vars="State", var_name="Year", value_name="Employment")
```

```python
df_employment.head()
```

```python
df_employment = df_employment.dropna()
```

```python
sorted(df_employment["State"].unique())
```

```python
df_employment.info()
```

```python
df_employment["Year"] = df_employment["Year"].astype(int)
```

```python
df_employment["State"].nunique()
```

```python
df_employment = df_employment.reset_index(drop=True)
```

```python
df_employment = df_employment.sort_values(by=["State", "Year"])
```

```python
df_employment.head(20)
```

```python
df_employment = df_employment[df_employment["Year"] != 2025]
```

```python
df_employment = df_employment.reset_index(drop=True)
```

```python
df_employment.head(20)
```

```python
df_employment.tail(20)
```

```python
df_employment.to_csv("../data/cleaned/employment_clean.csv", index=False)
```
