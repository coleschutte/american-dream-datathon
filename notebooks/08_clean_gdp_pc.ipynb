```python
import pandas as pd
import numpy as np
```

```python
raw_gdp = pd.read_csv("../data/cleaned/gdp_clean.csv")
gdp_pc = raw_gdp.copy()

raw_population = pd.read_csv("../data/cleaned/population_cleaned.csv")
population = raw_population.copy()
```

```python
gdp_pc = gdp_pc.merge(population[['State', 'Year', 'Population']], on=['State', 'Year'], how='left')
gdp_pc["GDP"] = gdp_pc["GDP"] * 1000000 / gdp_pc["Population"]
gdp_pc = gdp_pc.drop('Population', axis=1)
```

```python
gdp_pc.head()
```

```python
gdp_pc.info()
```

```python
gdp_pc = gdp_pc[gdp_pc["Year"] >= 2008]
```

```python
gdp_pc.head()
```

```python
gdp_pc['GDP'] = gdp_pc['GDP'].round(2)
```

```python
gdp_pc.head()
```

```python
gdp_pc = gdp_pc.reset_index(drop=True)
```

```python
gdp_pc.head()
```

```python
gdp_pc.rename(columns={'GDP': 'gdp_pc'}, inplace=True)
```

```python
gdp_pc.head()
```

```python
print(f"Missing values:\n{gdp_pc.isnull().sum()}")
print(f"\nTotal rows: {len(gdp_pc)}")
print(f"\nGDP per capita range: ${gdp_pc['gdp_pc'].min():,.0f} to ${gdp_pc['gdp_pc'].max():,.0f}")
```

```python
gdp_pc.to_csv("../data/cleaned/gdp_pc_clean.csv", index=False)
```
