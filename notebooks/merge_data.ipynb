```python
import pandas as pd
import numpy as np
```

```python
df_gdp = pd.read_csv("../data/cleaned/gdp_clean.csv")
df_income = pd.read_csv("../data/cleaned/income_pc_clean.csv")
df_real_income = pd.read_csv("../data/cleaned/real_income_pc_clean.csv")
df_rpp = pd.read_csv("../data/cleaned/regional_price_parity_clean.csv")
df_employment = pd.read_csv("../data/cleaned/employment_clean.csv")
```

```python
df_gdp["Year"].unique()
```

```python
for i in range(1997,2008):
    df_gdp = df_gdp[df_gdp["Year"]!= i]
```

```python
df_income["Year"].unique()
```

```python
df = df_gdp.merge(df_income, on=["State", "Year"], how="outer")
df = df.merge(df_real_income, on=["State", "Year"], how="outer")
df = df.merge(df_rpp, on=["State", "Year"], how="outer")
df = df.merge(df_employment, on=["State", "Year"], how="outer")
```

```python
for i in range(1997,2008):
    df = df[df["Year"]!= i]
```
