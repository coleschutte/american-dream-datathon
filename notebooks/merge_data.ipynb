```python
# Load additional datasets to add to master_v1
df_gdp_pc = pd.read_csv("../data/cleaned/gdp_pc_clean.csv")
df_employment_pc = pd.read_csv("../data/cleaned/employment_pc_clean.csv")

master = pd.read_csv("../data/processed/master_dataset_v1.csv")

df_gdp_pc.columns = ['state', 'year', 'gdp_pc']
df_employment_pc.columns = ['state', 'year', 'employment_pc']

master = master.merge(df_gdp_pc, on=['state', 'year'], how='left')
master = master.merge(df_employment_pc, on=['state', 'year'], how='left')

master.to_csv("../data/processed/master_dataset_v1.csv", index=False)

print(f"Master dataset shape: {master.shape}")
print(f"\nColumns: {list(master.columns)}")
```

```python
# Create organized master_dataset_v2
master_v2 = master.copy()

# 1. Geographic/Time identifiers
# 2. Population (base metric)
# 3. Economic output (GDP metrics)
# 4. Employment metrics
# 5. Income metrics
# 6. Cost of living adjustment

column_order = [
    'state', 'year', 'population',
    'gdp', 'gdp_pc',
    'employment', 'employment_pc',
    'income_pc', 'real_income_pc',
    'rpp'
]

master_v2 = master_v2[column_order]

# Check for any missing values
print("Missing values per column:")
print(master_v2.isnull().sum())
print("\n" + "="*60)
print(f"Dataset shape: {master_v2.shape}")
print(f"States: {master_v2['state'].nunique()}")
print(f"Year range: {master_v2['year'].min()}-{master_v2['year'].max()}")
print("="*60 + "\n")

# Display sample data
print("Sample data (first 5 rows):")
print(master_v2.head())
print("\nSample data (random state):")
print(master_v2[master_v2['state'] == 'California'].head())
```

```python
# Summary statistics by column
print("\nColumn Summary Statistics:")
print("="*60)
print(master_v2.describe().round(2))
print("\nData types:")
print(master_v2.dtypes)
```

```python
master_v2.to_csv("../data/processed/master_dataset_v2.csv", index=False)
```
