```python
import numpy as np
import pandas as pd
```

```python
raw_employment = pd.read_csv("../data/cleaned/employment_clean.csv")
employment_pc = raw_employment.copy()

raw_population = pd.read_csv("../data/cleaned/population_cleaned.csv")
population = raw_population.copy()
```

```python
employment_pc.head(20)
```

```python
population.head(20)
```

```python
employment_pc = employment_pc.merge(population[['State', 'Year', 'Population']], on=['State', 'Year'], how='left')
employment_pc["Employment"] = employment_pc["Employment"] / employment_pc["Population"]
employment_pc = employment_pc.drop('Population', axis=1)
```

```python
employment_pc.head(20)
```

```python
employment_pc = employment_pc[employment_pc["Year"] >= 2008]
```

```python
employment_pc.tail(20)
```

```python
employment_pc = employment_pc.reset_index(drop=True)
```

```python
employment_pc.rename(columns={'Employment': 'employment_pc'}, inplace=True)
```

```python
employment_pc.head(20)
```

```python
print(f"Missing values:\n{employment_pc.isnull().sum()}")
print(f"\nTotal rows: {len(employment_pc)}")
```

```python
employment_pc.to_csv("../data/cleaned/employment_pc_clean.csv", index=False)
```
