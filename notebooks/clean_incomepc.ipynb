```python
import pandas as pd
import numpy as np
```

```python
incomepc_raw = pd.read_csv("../data/raw/income_per_capita.csv")
```

```python
incomepc = incomepc_raw.copy()
```

```python
incomepc.head()
```

```python
incomepc = incomepc.drop(columns=["GeoFIPS"])
```

```python
incomepc.head()
```

```python
for i in range(1929, 1997):
    incomepc = incomepc.drop(columns=[str(i)])
```

```python
incomepc.head()
```

```python
incomepc = incomepc[incomepc["GeoName"] != "United States"]
```

```python
incomepc = incomepc.rename(columns={"GeoName": "State"})
```

```python
incomepc.info()
```

```python
incomepc = incomepc.melt(id_vars="State", var_name="Year", value_name="Income_pc")
```

```python
incomepc.head()
```

```python
incomepc = incomepc.sort_values(by=["State", "Year"])
```

```python
incomepc.head()
```

```python
incomepc = incomepc.reset_index(drop=True)
```

```python
incomepc.tail()
```

```python
incomepc = incomepc.dropna()
```

```python
incomepc.tail()
```

```python
incomepc.isna().sum()
```

```python
incomepc["State"].unique()
```

```python
regions = [
    "Far West *", "Great Lakes", "Mideast", "New England",
    "Plains", "Rocky Mountain", "Southeast", "Southwest"
]

incomepc = incomepc[~incomepc["State"].isin(regions)]
```

```python
incomepc["State"].unique()
```

```python
incomepc["State"] = incomepc["State"].str.replace(" *", "", regex=False)
```

```python
incomepc["State"].unique()
```

```python
incomepc.head(20)
```

```python
incomepc.info()
```

```python
incomepc["Year"] = incomepc["Year"].astype(int)
```

```python
incomepc.info()
```

```python
incomepc = incomepc.reset_index(drop=True)
```

```python
incomepc.info()
```

```python
incomepc["State"].value_counts()
```

```python
incomepc.to_csv("../data/cleaned/incomepc_clean.csv", index=False)
```
