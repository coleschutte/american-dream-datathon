```python
import pandas as pd
import numpy as np
```

```python
rpp_raw = pd.read_csv("../data/raw/regional_price_parity.csv")
```

```python
rpp = rpp_raw.copy()
```

```python
rpp.head()
```

```python
rpp = rpp.drop(columns="GeoFIPS")
```

```python
rpp.head()
```

```python
rpp.info()
```

```python
rpp.isna().sum()
```

```python
rpp.tail()
```

```python
rpp = rpp.dropna()
```

```python
rpp = rpp.rename(columns={"GeoName" : "State"})
```

```python
rpp.head()
```

```python
rpp = rpp[rpp["State"] != "United States"]
```

```python
rpp.head()
```

```python
rpp = rpp.melt(id_vars="State", var_name="Year", value_name="Regional Price Parity")
```

```python
rpp.head()
```

```python
rpp["State"].unique()
```

```python
rpp = rpp.sort_values(by=["State", "Year"])
```

```python
rpp.head()
```

```python
rpp = rpp.reset_index(drop=True)
```

```python
rpp.head()
```

```python
rpp.tail()
```

```python
rpp.info()
```

```python
rpp["Year"] = rpp["Year"].astype(int)
```

```python
rpp = rpp.rename(columns={"Regional Price Parity" : "RPP"})
```
