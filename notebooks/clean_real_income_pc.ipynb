```python
import pandas as pd
import numpy as np
```

```python
rincomepc_raw = pd.read_csv("../data/raw/real_income_per_capita.csv")
```

```python
rincomepc = rincomepc_raw.copy()
```

```python
rincomepc.head()
```

```python
rincomepc.info()
```

```python
rincomepc = rincomepc.drop(columns="GeoFIPS")
```

```python
rincomepc = rincomepc.dropna()
```

```python
rincomepc.head()
```

```python
rincomepc = rincomepc[rincomepc["GeoName"] != "United States"]
```

```python
rincomepc = rincomepc.rename(columns={"GeoName":"State"})
```

```python
rincomepc.head()
```

```python
rincomepc = rincomepc.melt(id_vars="State", var_name="Year", value_name="Real_Income_pc")
```

```python
rincomepc.head()
```

```python
rincomepc.tail()
```

```python
rincomepc.isna().sum()
```

```python
rincomepc["State"].unique()
```

```python
rincomepc = rincomepc.sort_values(by=["State", "Year"])
```

```python
rincomepc.head()
```

```python
rincomepc = rincomepc.reset_index(drop=True)
```

```python
rincomepc.head()
```

```python
rincomepc.info()
```

```python
rincomepc["Year"] = rincomepc["Year"].astype(int)
```

```python
rincomepc.head()
```

```python
rincomepc.tail()
```

```python
rincomepc.info()
```

```python
rincomepc.to_csv("../data/cleaned/real_income_pc_clean.csv", index=False)
```
