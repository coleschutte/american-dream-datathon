```python
import pandas as pd
import numpy as np
```

```python
raw_pop = pd.read_csv('../data/raw/population.csv')
pop = raw_pop.copy()
```

```python
pop.head()
```

```python
pop = pop.drop(columns=['GeoFIPS'])
```

```python
pop = pop.rename(columns={'GeoName': 'State'})
```

```python
pop.head()
```

```python
cols_to_drop = [str(i) for i in range(1929, 2008)]
pop = pop.drop(columns=cols_to_drop)
```

```python
pop.head()
```

```python
pop = pop[pop['State'] != 'United States']
```

```python
pop.head()
```

```python
pop['State'].unique
```

```python
pop = pop.dropna()
```

```python
pop["State"] = pop["State"].str.replace(" *", "", regex=False)
```

```python
pop.head()
```

```python
regions = [
    "Far West", "Great Lakes", "Mideast", "New England",
    "Plains", "Rocky Mountain", "Southeast", "Southwest"
]

pop = pop[~pop["State"].isin(regions)]
```

```python
pop.head()
```

```python
pop["State"].unique
```

```python
pop.info()
```

```python
pop = pop.melt(id_vars="State", var_name="Year", value_name="Population")
```

```python
pop.head()
```

```python
pop = pop.sort_values(by=["State", "Year"])
```

```python
pop['Year'] = pop['Year'].astype(int)
pop['Population'] = pop['Population'].astype(int)
```

```python
pop.head()
```

```python
pop.info()
```

```python
pop.head(20)
```

```python
pop = pop.reset_index(drop=True)
```

```python
pop.head(20)
```

```python
pop = pop[pop["Year"] != 2025]
```

```python
pop.head(20)
```

```python
pop.to_csv('../data/cleaned/population_cleaned.csv', index=False)
```
